### Q4. string.punctuation 을 이용해서 특수문자를 제거해주세요.

#### Q7. 입력문서가 {fast, furious, fun} 을 주요 단어로 가질때, 이 문서는 얼마의 확률로 어떤 문서로 분류가 될까요? ( 계산과정을 채워주세요)

```python
doc = np.array(doc)        # dataframe을 np-array로 변환해줍니다.
                         # return값은 np-matrix가 아니라 np-array입니다.
doc
```

####  P(Y=comedy), P(Y=action) 계산하기

```python
# P(Y=comedy)
p_comedy = sum(doc[:,0]==1) / len(doc)

# P(Y=action)
p_action = sum(doc[:,0]==0) / len(doc)
```

```python
print('p_comedy : ', p_comedy)
print('p_action : ', p_action)
```

#### P(fast=1|comedy=1), P(furious=1|comedy=1), P(fun=1|comedy=1) 계산하기
* 참고 : 문서에 두번 등장한 단어 주의

```python
# P(fast=1|comedy=1)
p_comedy_fast = sum((doc[:,1]>=1) & (doc[:,0]==1)) / sum(doc[:,0]==1)

# P(furious=1|comedy=1)
p_comedy_furious = sum((doc[:,2]>=1) & (doc[:,0]==1)) / sum(doc[:,0]==1)

# P(fun=1|comedy=1)
p_comedy_fun = sum((doc[:,3]>=1) & (doc[:,0]==1)) / sum(doc[:,0]==1)
```

```python
print('p_comedy_fast : ' , p_comedy_fast)
print('p_comedy_furious : ' , p_comedy_furious)
print('p_comedy_fun : ' , p_comedy_fun)
```

#### P(fast=1|action=1), P(furious=1|action=1), P(fun=1|action=1) 계산하기
* 참고 : 문서에 두번 등장한 단어 주의

```python
# P(fast=1|action=1)
p_action_fast = sum((doc[:,1]>=1) & (doc[:,0]==0)) / sum(doc[:,0]==0)

# P(furious=1|action=1)
p_action_furious = sum((doc[:,2]>=1) & (doc[:,0]==0)) / sum(doc[:,0]==0)

# P(fun=1|action=1)
p_action_fun = sum((doc[:,3]>=1) & (doc[:,0]==0)) / sum(doc[:,0]==0)
```

```python
print('p_action_fast : ' , p_action_fast)
print('p_action_furious : ' , p_action_furious)
print('p_action_fun : ' , p_action_fun)
```

#### P(Y = comedy| X = fast, furious, fun) , P(Y = action=1| X = fast, furious, fun) 값 계산하기

```python
# P(X|comedy)와 P(X|action) 계산
p_x_given_comedy = p_comedy_fast * p_comedy_furious * p_comedy_fun
p_x_given_action = p_action_fast * p_action_furious * p_action_fun

# P(X) 계산 (전체 확률의 법칙을 사용)
p_x = p_x_given_comedy * p_comedy + p_x_given_action * p_action

# P(Y = comedy| X = fast, furious, fun) 계산
proba_comedy = (p_x_given_comedy * p_comedy) / p_x

# P(Y = action=1| X = fast, furious, fun) 계산
proba_action = (p_x_given_action * p_action) / p_x
```

```python
print('proba_comedy', proba_comedy)
print('proba_action', proba_action)
```

Q7. 어떤 문제점을 발견할수 있나요? 문제점을 해결하기 위한 방법으로는 어떤게 있을까요?

**문제점**
- **단어의 빈도 무시**: 기본 나이브 베이즈 모델은 단어의 등장 횟수를 고려하지 않고, 단순히 단어의 존재 여부만을 고려한다. 이는 문서의 맥락이나 단어의 중요성을 충분히 반영하지 못할 수 있다.

- **단어의 상호 독립 가정**: 나이브 베이즈 모델은 모든 특성(단어)이 서로 독립적이라고 가정한다. 하지만 실제 텍스트 데이터에서는 단어들 사이에 종종 강한 상관관계가 있을 수 있다. 예를 들어, 'fast'와 'furious'는 종종 함께 등장할 가능성이 높다.

- **희귀 단어 문제**: 특정 단어가 훈련 데이터에서 매우 드물게 등장하거나 전혀 등장하지 않는 경우, 이 단어가 테스트 문서에 등장할 때 모델이 해당 단어를 처리하는 방식에 문제가 발생할 수 있다.

**해결 방안**
- **TF-IDF 사용**: 단어의 빈도수와 함께 문서 전체에서의 단어의 중요성을 고려하는 TF-IDF 점수를 사용하여 단어의 가중치를 조정할 수 있다. 이는 단순한 단어의 존재 여부보다 더 많은 정보를 모델에 제공한다.

- **스무딩 기법 적용**: 특정 단어가 훈련 데이터에 없을 경우 확률이 0이 되는 것을 방지하기 위해, 모든 단어에 작은 확률 값을 추가하는 스무딩 기법을 적용할 수 있습니다. 이는 모델이 희귀 단어를 처리하는 능력을 향상시킨다.

- **특성 선택 및 차원 축소**: 상호 의존적인 특성을 줄이기 위해, 특성 선택 기법을 사용하여 가장 유용한 특성만을 모델에 포함시킬 수 있다. 또한, 차원 축소 기법을 통해 데이터의 차원을 줄이면서 중요한 정보를 유지할 수 있다.

- **다른 모델과의 결합**: 나이브 베이즈 모델만으로 충분하지 않은 경우, 다른 분류 알고리즘과 함께 사용하여 성능을 향상시킬 수 있다. 앙상블 학습 방법을 적용하면 다양한 모델의 장점을 결합할 수 있다.
