```python
## Unknown이 없는 열은 모두 1과 0으로 바꾼 후 진행

df['Attrition_Flag'] = np.where(df['Attrition_Flag'] == 'Existing Customer', 0, 1)
df['Gender'] = np.where(df['Gender']=="F",0, 1)
```

```python
## Unknown이 있는 열은 모두 제거 한 데이터로 진행

train_df = df.copy()

train_df = pd.concat([ train_df, pd.get_dummies( train_df['Education_Level']).drop(columns=['Unknown'])], axis=1)
train_df = pd.concat([ train_df, pd.get_dummies( train_df['Income_Category']).drop(columns=['Unknown'])], axis=1)
train_df = pd.concat([ train_df, pd.get_dummies( train_df['Marital_Status']).drop(columns=['Unknown'])], axis=1)
train_df = pd.concat([ train_df, pd.get_dummies( train_df['Card_Category'])], axis=1)

 ## get_dummies를 이용해 더미화 시킨 후 이후 해당 열 제거
train_df.drop(columns = ['Education_Level', 'Income_Category', 'Marital_Status', 'Card_Category'], inplace=True)
train_df
```

```python
train_df.isna().sum().sum()

## 최종 결측치 - 0
```

```python
from sklearn.model_selection import train_test_split

X = train_df.drop(columns=['Attrition_Flag'])
y = train_df['Attrition_Flag']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
import time
```

```python
## 모형 적합

### RandomForest
rf_clf = RandomForestClassifier()
start = time.time()
rf_clf.fit(X_train, y_train)
end = time.time()

## 결과 파악
y_pred = rf_clf.predict(X_test)

print("소요 시간 : ", end - start)
print("Accuracy : ", accuracy_score(y_test, y_pred))
print("Recall score: ", recall_score(y_test, y_pred))
```

```python
## 모형 적합

### AdaBoost
ada_clf = AdaBoostClassifier()
start = time.time()
ada_clf.fit(X_train, y_train)
end = time.time()

## 결과 파악
y_pred = ada_clf.predict(X_test)

print("소요 시간 : ", end - start)
print("Accuracy : ", accuracy_score(y_test, y_pred))
print("Recall score: ", recall_score(y_test, y_pred))
```

```python
## 모형 적합

### XGBoost
xgb_clf = XGBClassifier()
start = time.time()
xgb_clf.fit(X_train, y_train)
end = time.time()

## 결과 파악
y_pred = xgb_clf.predict(X_test)

print("소요 시간 : ", end - start)
print("Accuracy : ", accuracy_score(y_test, y_pred))
print("Recall score: ", recall_score(y_test, y_pred))
```
